```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=apple%20mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
r=requests.get(url)
soup=BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python

container=soup.find_all('div', class_="_2pi5LC col-12-12")
```

```python
prices=[]
product=[]
ratings=[]
```

```python
# getting names
a=soup.find_all('div',class_='_4rR01T')

# b=list(a)
for n in a:
    ns=(n.text)
    product.append(ns)
```

```python
print(product)
```

```python
# getting price
price=soup.find_all('div', class_="_30jeq3 _1_WHN1")

for p in price:
    ps=(p.text)
    prices.append(ps)
     
```

```python
print(prices)
```

```python
#getting the ratings
rating=soup.find_all('div', class_="_3LWZlK")

for r in rating:
    rs=(r.text)
    ratings.append(rs)

```

```python
print(ratings)
```
